```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [8,8]
plt.rcParams.update({'font.size':18})

x =3 #True slope
a = np.arange(-2,2,0.25)
a = a.reshape(-1,1)

b = x*a + np.random.randn(*a.shape) # with noise

plt.plot(a, x*a, color='k', linewidth=2, label='True line')
plt.plot(a, b, 'x',  color='r', markersize=10, label= 'Data')

U, S, VT = np.linalg.svd(a, full_matrices=False)
# least square fit
xtilde = VT.T @ np.linalg.inv(np.diag(S)) @ U.T @ b 

plt.plot(a, xtilde * a, '--', color='b', linewidth=4, label="Regresssion linear")
plt.xlabel('a')
plt.ylabel('b')
plt.grid(linestyle='--')

plt.legend()
plt.show()
```
